```python
import pandas as pd

w1 = pd.read_excel('LY Case Study Quantitative_11.11.24.xlsx',sheet_name='W-1')

w = pd.read_excel('LY Case Study Quantitative_11.11.24.xlsx',sheet_name='W')
```

```python
country_columns_w1 = ['France', 'United Kingdom', 'USA (tax free)', 'Japan', 'Hong Kong', 'China',
                   'Korea', 'Korea DF', 'Hainan DF', 'United Arab Emirates', 'Australia', 
                   'Canada', 'Macao', 'Mexico', 'Saudi Arabia', 'Singapore', 'Thailand', 'Taiwan']

w1_unpivot = pd.melt(w1 , id_vars=['Brand' , 'Name' , 'Category', 'SKU', 'URL'],
value_vars=country_columns_w1,
var_name='Country', value_name='Price')
```

```python
w1_unpivot.head()
```

```python
w.columns
```

```python
country_columns_w = ['France', 'United Kingdom', 'USA (tax free)', 'Japan', 'Hong Kong', 'China',
                   'Korea', 'Korea DF', 'Hainan DF', 'United Arab Emirates', 'Australia', 
                   'Canada', 'Macao', 'Mexico', 'Saudi Arabia', 'Singapore', 'Thailand', 'Taiwan']

w_unpivot = pd.melt(w , id_vars=['Brand' , 'Name' , 'Category', 'SKU'],
value_vars=country_columns_w,
var_name='Country', value_name='Price')
```

```python
w_unpivot.head()
```

```python
price_difff = w_unpivot.groupby('SKU')['Price'].agg('min','max')
```

```python
price_difff.head(50)
```

```python
df_combined = pd.merge(w_unpivot , w1_unpivot , on=['SKU','Country'], how='outer')
```

```python
df_combined
```

```python
df = df_combined.drop(columns=['Category_x','Category_y','URL'])

df
```

```python
df.to_excel('processed_data.xlsx')
```

```python
df_new['Price_Diff'] = df_new['Price_y']-df_new['Price_x']

df_new['Price Pct Change']= (df_new['Price_Diff']/df_new['Price_x'])*100

df_new.fillna(0,inplace=True)

df_new

```

```python
df_new.rename(columns={
    'Price_x':'Price_w',
              'Name_x':'Name_w',
              'Brand_x':'Brand_w',
              'Price_y':'Price_w1',
              'Name_y':'Name_w1',
              'Brand_y':'Brand_w1'
              },inplace=True)

df_new.head()
```

```python
df_processed = df_new
```

```python
df_processed.to_excel('Processed_data.xlsx',index=False)
```
